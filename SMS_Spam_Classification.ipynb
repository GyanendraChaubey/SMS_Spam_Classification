```python
import pandas as pd
```

```python
dataset=pd.read_csv('https://raw.githubusercontent.com/GyanendraChaubey/SMS_Spam_Classification/main/SMS_Spam.csv',sep=',',header=None,error_bad_lines=False)
```

```python
dataset
```

```python
dataset.columns=['lables','message']
```

```python
#dataset.to_csv('SMS_Spam.csv')
```

```python
dataset.head()
```

```python
#data cleaning
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
from  nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i  in range(0, len(dataset)):
  review=re.sub('[^a-zA-Z0-9]',' ',dataset['message'][i])
  review=review.lower()
  review=review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
##Train and Test Split
```

```python
##independent and dependent features
y=pd.get_dummies(dataset['lables'],drop_first=True)
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(corpus,y,test_size=0.3)
```

#Bag of Words

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
x_train = cv.fit_transform(x_train).toarray()
```

```python
import pickle
pickle.dump(cv,open('BOW.pkl','wb'))
```

```python
x_train
```

```python
cv.vocabulary_
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
x_test=cv.transform(x_test)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
#Save the model
import pickle
pickle.dump(clf,open('model.pkl','wb'))
```

#Tf_idf

```python
# Creating the tfidf model
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf = TfidfVectorizer(max_features=2500)
x_train = tf_idf.fit_transform(x_train).toarray()
```

```python
import pickle
pickle.dump(tf_idf,open('tf_idf.pkl','wb'))
```

```python
x_train
```

```python
tf_idf.vocabulary_
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
import pickle
tfidf=pickle.load(open('tf_idf.pkl','rb'))
```

```python
x_test=tfidf.transform(x_test)
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
print(classification_report(y_test,y_pred))
```
